__Income Statement__

```python
import tabula
import pandas as pd
import numpy as np

pages = '4'
columns = [40,410,490]
area = (20, 0, 75, 100)
df = tabula.read_pdf("AAPL - Q1-2020.pdf", stream=True, guess=False, pages=pages, 
                     area=area, relative_area=True, columns=columns)

df = df[0].drop(["Unnamed: 0"], axis=1)
df = df.drop([0])

df.rename(columns={'Unnamed: 1': 'Income Statement', 
                   'December 28,': 'Q1 2020', 
                   'December 29,': 'Q1 2019'}, inplace=True)

df.replace(np.nan, '', inplace=True)
df.replace('\$', '', inplace=True, regex=True)
df
```
